```python
#1. (Rekurzija) Sledeće zadatke uraditi rekurzivno:
#b. Napisati rekurzivnu funkciju koja racuna Fibonačijev broj za koji važi, n- prirodan broj veci od 1; F0 = 0, F1 = 1, Fn = Fn−1 + Fn−2

def fibonacijev_br(n):
    if n <= 0:
        return 0
    elif n == 1:
        return 1
    else:
        return fibonacijev_br(n-1) + fibonacijev_br(n-2)

# Testiranje za n = 5
print(fibonacijev_br(5)) # Output: 5, jer je F₅ = F₄ + F₃ = 3 + 2 = 5
```

```python
#Napisati rekurzivnu funkcija koja pretvara prirodan dekadni broj u birani broj.

def dekadni_u_binarni(n):
    if n == 0:
        return "0"
    elif n == 1:
        return "1"
    else:
        return dekadni_u_binarni(n // 2) + str(n % 2)

# Test sa 10 brojeva
for i in range(10):
    print(f"{i} u binarnom zapisu je: {dekadni_u_binarni(i)}")


#Binarni zapis se dobija tako što se broj dijeli sa 2, pamti se ostatak, a zatim se količnik ponovo dijeli sa 2, i tako dalje dok količnik ne postane 0 ili 1.
#Ostaci, čitani od poslednjeg ka prvom, daju binarni zapis.
```

```python
#Napisati rekurzivnu funkciju koja vraća poslednje veliko slovo zadatog stringa.
#Input: “Neka Ulazna Recenica” Output: “R”
def posl_veliko(s):
    if not s:
        return None
    if s[-1].isupper():
        return s[-1]
    return posl_veliko(s[:-1])

# Test funkcije
test = ["Neka Ulazna Recenica"]
for string in test:
    result = posl_veliko(string)
    print(f"Posljednje veliko slovo u '{string}' je: {result}")
```

```python
#Napisati rekurzivnu funkciju koja provjerava koliko zadati broj ima parnih cifara.

def parne_cifre(n):
    if n == 0:
        return 1

    if n < 0:
        n = abs(n)

    if n < 10:
        return 1 if n % 2 == 0 else 0
    else:
        last_digit = n % 10
        count_from_rest = parne_cifre(n // 10)
        if last_digit % 2 == 0:
            return count_from_rest + 1
        else:
            return count_from_rest

# Test funkcije
test_numbers = [123, 2468, 13579, 1234567890]
for num in test_numbers:
    print(f"Broj {num} ima {parne_cifre(num)} parnih cifara.")

#Funkcija "rastavlja" broj na cifre i broji one koje su parne, krećući se od posljednje cifre prema prvoj.
```

```python
#Napisati funkciju koja na osnovu ulazne liste l1, kreira i štampa novu
#listu l2, a koja se sastoji od parnih elemenata iz liste l1 koje je potrebno kvadrirati
#Input: 5 -> 4 -> 3 -> 8 ->2 Output: 16 -> 64 -> 4
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class LinkedList:
    def __init__(self):
        self.head = None

    def append(self, data):
        new_node = Node(data)
        if not self.head:
            self.head = new_node
            return

        current = self.head
        while current.next:
            current = current.next
        current.next = new_node

    def print_list(self):
        current = self.head
        elements = []
        while current:
            elements.append(str(current.data))
            current = current.next
        print(" -> ".join(elements))

    def create_squared_even_list(self):
        result = LinkedList()
        current = self.head

        while current:
            if current.data % 2 == 0:  # označava da je element paran
                result.append(current.data ** 2)  #kvadrira taj parni element
            current = current.next

        return result

# Test funkcije
original_list = LinkedList()
for num in [5, 4, 3, 8, 2]:
    original_list.append(num)

print("Originalna lista:")
original_list.print_list()

squared_even_list = original_list.create_squared_even_list()
print("Lista kvadriranih parnih elemenata:")
squared_even_list.print_list()
```

```python
#Napisati funkciju koja uklanja najveći element iz liste (podsjetnik da imate implementiranu funkciju koja briše element na osnovu zadate vrijednosti – nedelja 2)
#Input: 5 -> 4 -> 3 -> 7 -> 5 -> 1 Output: 5 -> 4 -> 3 -> 5 -> 1

class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class LinkedList:
    def __init__(self):
        self.head = None

    def append(self, data):
        new_node = Node(data)
        if not self.head:
            self.head = new_node
            return

        current = self.head
        while current.next:
            current = current.next
        current.next = new_node

    def print_list(self):
        current = self.head
        elements = []
        while current:
            elements.append(str(current.data))
            current = current.next
        print(" -> ".join(elements))

    #uklanja element sa najvećom vrijednošću
    def remove_max(self):
        if not self.head:
            return

        # Pronalazi element sa najvećom vrijednošću
        max_val = self.head.data
        current = self.head
        while current:
            if current.data > max_val:
                max_val = current.data
            current = current.next

        if self.head.data == max_val:
            self.head = self.head.next
            return

        current = self.head
        while current.next and current.next.data != max_val:
            current = current.next

        if current.next:
            current.next = current.next.next

# Test funkcije
test_list = LinkedList()
for num in [5, 4, 3, 7, 5, 1]:
    test_list.append(num)

print("Originalna lista:")
test_list.print_list()

test_list.remove_max()
print("Lista nakon uklanjanja najvećeg elementa:")
test_list.print_list()
```

```python
#Dvostruko olančane liste: Brisanje poslednja dva elementa iz dv. olančane liste.
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
        self.prev = None

class DoublyLinkedList:
    def __init__(self):
        self.head = None
        self.tail = None

    def append(self, data):
        new_node = Node(data)
        if not self.head:
            self.head = new_node
            self.tail = new_node
            return

        new_node.prev = self.tail
        self.tail.next = new_node
        self.tail = new_node

    def print_list(self):
        current = self.head
        elements = []
        while current:
            elements.append(str(current.data))
            current = current.next
        print(" -> ".join(elements))

    def remove_last_two(self):
        # Ako lista ima manje od dva elementa
        if not self.head or not self.head.next:
            self.head = None
            self.tail = None
            return

        # Ako lista ima tačno dva elementa
        if not self.head.next.next:
            self.head = None
            self.tail = None
            return

        # Brisanje poslednja dva elementa
        self.tail = self.tail.prev.prev
        self.tail.next = None

# Test funkcije
test_list = DoublyLinkedList()
for num in [1, 2, 3, 4, 5]:
    test_list.append(num)

print("Originalna lista:")
test_list.print_list()

test_list.remove_last_two()
print("Lista nakon brisanja poslednja dva elementa:")
test_list.print_list()
```

```python
#U čvorovima sortirane dvostruko povezane liste nalaze se zapisi o knjigama: šifra knjige (int), naziv knjige (string), šifra autora (int) i cijena knjige
#(float). Lista je uzlazno sortirana prema cijeni knjige. Potrebno je napisati funkciju koja će iz liste izbaciti knjige s najvišom i najmanjom cijenom.
#Napomena: više knjiga može imati istu cijenu

class KnjigaNode:
    def __init__(self, sifra, naziv, sifra_autora, cijena):
        self.knjiga = {
            'sifra': sifra,
            'naziv': naziv,
            'sifra_autora': sifra_autora,
            'cijena': cijena
        }
        self.next = None
        self.prev = None

class DoublyLinkedListKnjige:
    def __init__(self):
        self.head = None
        self.tail = None

    def append(self, knjiga_node):
        if not self.head:
            self.head = knjiga_node
            self.tail = knjiga_node
            knjiga_node.prev = None
        else:
            # Dodaj na kraj i sortiraj po cijeni
            knjiga_node.prev = self.tail
            self.tail.next = knjiga_node
            self.tail = knjiga_node
            self.sort_by_price()

    def sort_by_price(self):
        # Jednostavno sortiranje liste po cijeni
        if not self.head:
            return

        swapped = True
        while swapped:
            swapped = False
            current = self.head

            while current.next:
                if current.knjiga['cijena'] > current.next.knjiga['cijena']:
                    # Zamijeni podatke u čvorovima
                    current.knjiga, current.next.knjiga = current.next.knjiga, current.knjiga
                    swapped = True
                current = current.next

    def remove_min_max_price(self):
        if not self.head or self.head == self.tail:
            return

        # Pronađi minimalnu i maksimalnu cijenu
        min_price = self.head.knjiga['cijena']
        max_price = self.tail.knjiga['cijena']

        # Ukloni sve knjige s minimalnom cijenom
        while self.head and self.head.knjiga['cijena'] == min_price:
            self.head = self.head.next
            if self.head:
                self.head.prev = None

        # Ako je lista prazna nakon uklanjanja minimalnih
        if not self.head:
            self.tail = None
            return

        # Ukloni sve knjige s maksimalnom cijenom
        while self.tail and self.tail.knjiga['cijena'] == max_price:
            self.tail = self.tail.prev
            if self.tail:
                self.tail.next = None

        # Provjeri srednje elemente s minimalnom cijenom
        current = self.head
        while current and current.next:
            if current.next.knjiga['cijena'] == min_price:
                current.next = current.next.next
                if current.next:
                    current.next.prev = current
                else:
                    self.tail = current
            else:
                current = current.next

        # Provjeri srednje elemente s maksimalnom cijenom
        current = self.head
        while current and current.next:
            if current.next.knjiga['cijena'] == max_price:
                current.next = current.next.next
                if current.next:
                    current.next.prev = current
                else:
                    self.tail = current
            else:
                current = current.next

    def print_all_books(self):
        current = self.head
        while current:
            print(f"Naziv: {current.knjiga['naziv']}, Cijena: {current.knjiga['cijena']}")
            current = current.next

# Test funkcije
knjige = DoublyLinkedListKnjige()

# Dodaj knjige
knjige.append(KnjigaNode(101, "Rat i mir", 1, 25.99))
knjige.append(KnjigaNode(102, "Ana Karenjina", 1, 19.99))
knjige.append(KnjigaNode(103, "Zločin i kazna", 2, 22.50))
knjige.append(KnjigaNode(104, "Braća Karamazovi", 2, 29.99))
knjige.append(KnjigaNode(105, "Idiot", 2, 19.99))
knjige.append(KnjigaNode(106, "Gospodar prstenova", 3, 35.00))

print("Sve knjige prije uklanjanja (sortirane po cijeni):")
knjige.print_all_books()

knjige.remove_min_max_price()

print("\nKnjige nakon uklanjanja onih s najvišom i najmanjom cijenom:")
knjige.print_all_books()
```
